## Barplot

## pie plot 

```python
x =np.array([1,2,3,4])
lable =np.array(['Co2', 'N2', 'o2', 'he'])
plt.pie(x, labels=lable, startangle=90)
plt.show()
```

```python
x =np.array([1,2,3,4])
lable =np.array(['Co2', 'N2', 'o2', 'he'])
myexplod = np.array([0.2,0,0,0])
plt.pie(x, labels=lable, startangle=90, explode=myexplod)
plt.show()
```

```python
x =np.array([1,2,3,4])
lable =np.array(['Co2', 'N2', 'o2', 'he'])
myexplod = np.array([0.2,0,0,0])
plt.pie(x, labels=lable, startangle=90, explode=myexplod,shadow=True)
plt.show()
```

```python
x =np.array([1,2,3,4])
lable =np.array(['Co2', 'N2', 'o2', 'he'])
myexplod = np.array([0,0.2,0,0])
c=np.array(['red', 'green', 'black','pink'])
plt.pie(x, labels=lable, startangle=90, explode=myexplod,shadow=True,colors= c )
plt.show()
```

```python
x =np.array([1,2,3,4])
lable =np.array(['Co2', 'N2', 'o2', 'he'])
myexplod = np.array([0,0.2,0,0])
c=np.array(['red', 'green', 'black','pink'])
plt.pie(x, labels=lable, startangle=90, explode=myexplod,shadow=True,colors= c,autopct='% 1.f%%' )
plt.show()
```

```python
x =np.array([1,2,3,4])
lable =np.array(['Co2', 'N2', 'o2', 'he'])
myexplod = np.array([0,0.2,0,0])
c=np.array(['red', 'green', 'black','pink'])
plt.pie(x, labels=lable, startangle=90, explode=myexplod,shadow=True,colors= c,autopct='% 1.f%%' )
plt.legend("My pie chart ")
plt.show() 
```

```python
x =np.array([1,2,3,4])
lable =np.array(['Co2', 'N2', 'o2', 'he'])
myexplod = np.array([0,0.2,0,0])
c=np.array(['red', 'green', 'black','pink'])
plt.pie(x, labels=lable, startangle=90, explode=myexplod,shadow=True,colors= c,autopct='% 1.f%%' )
plt.legend()
plt.show()
```
