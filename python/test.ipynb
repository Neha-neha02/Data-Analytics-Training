# Q1. What will be the output of the following Python code?


```python
def func(a, b=5, c=10):
    return a + b * c

print(func(2, c=3))
```

EXPLANATION:
In this function:

a = 2 (first positional argument)

b = 5 (default value, since it wasn’t provided)

c = 3 (overridden using keyword argument)

Now calculate:

a + b * c
= 2 + 5 * 3
= 2 + 15
= 17


Multiplication happens before addition (operator precedence), so the final output is:

17

# Q2. What is the scope of a variable declared inside a function?

```python
def my_func():
    x = 10   # local variable
    print("Inside function:", x)

my_func()

# print(x)  # This will give an error because x is not accessible outside
```

EXPLANATION:
* The variable x is created inside the function, so it has local scope.

* Local variables exist only within the function where they are defined.

When my_func() runs, it prints:

Inside function: 10


After the function finishes execution, the local variable x is destroyed.

If you uncomment print(x) outside the function, Python will raise a NameError because x is not defined in the global scope.

So, x is accessible only inside my_func(), not outside it.

# Q3. What does the following code return?

```python
def f(x, lst=[]):
    lst.append(x)
    return lst

print(f(1))
print(f(2))
```

EXPLANATION:
* The key concept here is mutable default arguments.

* The default value lst=[] is created only once, when the function is defined — not each time the function is called.

Since lists are mutable, changes to lst persist between function calls.

Step-by-step execution:

1️⃣ First call: f(1)

lst is the default empty list []

1 is appended → [1]

Output:

[1]


2️⃣ Second call: f(2)

The same list is reused (it is not reset)

Current list is [1]

2 is appended → [1, 2]

Output:

[1, 2]

Final Output:
[1]
[1, 2]


This happens because the default list is shared across function calls.

# Q 23. What does groupby() primarily perform?

```python
import pandas as pd

data = {"Department": ["IT", "HR", "IT", "HR"],
        "Salary": [50000, 40000, 60000, 45000]}

df = pd.DataFrame(data)

df.groupby("Department")["Salary"].mean()
```

EXPLANATION:
1. df.groupby("Department")

Groups the DataFrame by the values in the "Department" column.

Rows with the same department are grouped together.

Selecting "Salary" column

["Salary"] chooses the Salary column to perform operations on.

2. Calculating .mean()

Computes the average salary for each department:

IT: (50000 + 60000) / 2 = 55000

HR: (40000 + 45000) / 2 = 42500

Output:

Department
HR    42500.0
IT    55000.0
Name: Salary, dtype: float64

# Q24. What is the default join type in pd.merge()?

```python
import pandas as pd

df1 = pd.DataFrame({"id": [1, 2, 3], "name": ["A", "B", "C"]})
df2 = pd.DataFrame({"id": [2, 3, 4], "age": [20, 21, 22]})

pd.merge(df1, df2, on="id")
```

EXPLANATION:
1. pd.merge(df1, df2, on="id")

Combines df1 and df2 based on the common column "id".

By default, merge performs an inner join, keeping only rows with matching id values in both DataFrames.

Matching rows:

* id = 2 → present in both → merged row: id=2, name=B, age=20

* id = 3 → present in both → merged row: id=3, name=C, age=21

* id = 1 → only in df1 → excluded

* id = 4 → only in df2 → excluded

Output:

   id name  age
0   2    B   20
1   3    C   21
