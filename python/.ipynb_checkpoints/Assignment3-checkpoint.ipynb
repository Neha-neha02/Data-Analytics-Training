```python
#1.DAta Cleaning & Preprocessing
# (A) Load the data set amd check for missing values

import pandas as pd
df=pd.read_csv("train.csv")
print(df.isnull().sum())
```

```python
#(B) Fill missing Age values with median or mean 
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
print(df['Age'].isnull().sum())
```

```python
#(C) Handle missing Embarked values by filling with the most common port
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
#(D) Drop unnecessary columns like name ,Ticket,Cabin
df=df.drop(columns=['Name','Ticket','Cabin'])
```

```python
import pandas as pd
df=pd.read_csv("train.csv")
df.head()
```

```python
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], kde=True)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# (B) plot the count of passengers by sex using a bar chart
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', data=df)
plt.title("Count of Passengers by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
#(C) show the distribution of Fare using a boxplot
plt.figure(figsize=(6,4))
sns.boxplot(y='Fare', data=df)
plt.title("Distribution of Fare")
plt.ylabel("Fare")
plt.show()
```

```python
#Q.3 Bivariate Analysis
#(A) Compare survival rate across sex using a countplot
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival Rate Across Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
#(B) Compare Pclass vs Fare using a boxplot
plt.figure(figsize=(6,4))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Pclass vs Fare")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
#(C) Visualize Age vs Fare using a scatter plot, color-coded by Survived
plt.figure(figsize=(7,5))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare (Color-coded by Survival)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
